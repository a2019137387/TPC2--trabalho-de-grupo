**Exercício 1: (1 valor) Ler o lucro total (Total Profit) de todos os meses e mostre-os através de um gráfico de linhas.**

```python
import pandas as pd
import matplotlib.pyplot as plt  

df = pd.read_csv("company_sales_data.csv") #ler csv

fig=plt.figure()
ax=fig.add_axes([0,0,1,1]) #aumentar os eixos a nível de largura e altura para uma melhor visualização do gráfico

profitList = df ['total_profit'].tolist() #criar a lista do profit
monthList  = df ['month_number'].tolist() #criar a lista dos meses
plt.plot(monthList, profitList, label = 'Total_Profit') #linha do gráfico
plt.xlabel('Month number', fontsize= 14) #etiqueta do x com tamanho 14
plt.ylabel('Profit in dollar', fontsize= 14) #etiqueta do y com tamanho 14 
plt.xticks(monthList) #o que fica na coordenada do x é a lista dos meses
plt.title('Company profit per month', fontsize=17) #título do gráfico com tamanho 17
plt.yticks([100000, 200000, 300000, 400000, 500000]) #o que fica na coordenada do y são os valores do profit
plt.show()
```

**Exercício 2: (1 valor) Ler o lucro total (Total Profit) de todos os meses e 
apresente-os num gráfico de linhas, com as características indicadas.**


```python
import pandas as pd
import matplotlib.pyplot as plt  

#igual ao gráfico anterior só se altera a nível estético

df = pd.read_csv("company_sales_data.csv")
profitList = df ['total_profit'].tolist()
monthList  = df ['month_number'].tolist()

fig=plt.figure()
ax=fig.add_axes([0,0,1,1]) #aumentar os eixos a nível de largura e altura para uma melhor visualização do gráfico

plt.plot(monthList, profitList, label = 'Profit data of last year', #criar a linha do gráfico e estética
      color='r', marker='o', markerfacecolor='k', #cor; estilo linha; espessura; marker - círculo; markerfacecolor - preto;
      linestyle='--', linewidth=3)
      
plt.xlabel('Month Number',fontsize= 14) #etiqueta
plt.ylabel('Profit in dollar',fontsize= 14) #etiqueta
plt.legend(loc='lower right') #onde fica a legenda "profit data of last year"
plt.title('Company Sales data of last year',fontsize= 17) #título do gráfico
plt.xticks(monthList) #o que fica no eixo do x
plt.yticks([100000, 200000, 300000, 400000, 500000]) #o que fica no eixo do y
plt.show()
```

**Exercício 3: (1 valor) Mostrar as vendas mensais de todos os produtos num gráfico de múltiplas linhas, ou seja, 
uma linha diferente para cada produto.**


```python
import pandas as pd
import matplotlib.pyplot as plt  

df = pd.read_csv("company_sales_data.csv") #ler csv

fig=plt.figure()
ax=fig.add_axes([0,0,1,1]) #aumentar os eixos a nível de largura e altura para uma melhor visualização do gráfico

#criar as listas dos produtos
monthList  = df ['month_number'].tolist() 
faceCremSalesData   = df ['facecream'].tolist()
faceWashSalesData   = df ['facewash'].tolist()
toothPasteSalesData = df ['toothpaste'].tolist()
bathingsoapSalesData   = df ['bathingsoap'].tolist()
shampooSalesData   = df ['shampoo'].tolist()
moisturizerSalesData = df ['moisturizer'].tolist()

#criar e personalizar a linha de cada produto: etiqueta; marker; espessura.
plt.plot(monthList, faceCremSalesData,   label = 'Face cream Sales Data', marker='o', linewidth=3)
plt.plot(monthList, faceWashSalesData,   label = 'Face Wash Sales Data',  marker='o', linewidth=3)
plt.plot(monthList, toothPasteSalesData, label = 'ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(monthList, bathingsoapSalesData, label = 'Bathing Soap Sales Data', marker='o', linewidth=3)
plt.plot(monthList, shampooSalesData, label = 'Shampoo Sales Data', marker='o', linewidth=3)
plt.plot(monthList, moisturizerSalesData, label = 'Moisturizer Sales Data', marker='o', linewidth=3)

#personalizar o gráfico de acordo com as coordenadas do x e do y
plt.xlabel('Month Number',fontsize= 14) #etiqueta do x
plt.ylabel('Sales units in number', fontsize= 14) #etiqueta do y
plt.legend(loc='upper left') #localização das legendas
plt.xticks(monthList) #o que fica no eixo do x
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000]) #o que fica no eixo do y
plt.title('Sales data',fontsize= 17) #título do gráfico
plt.show()

 
 
```

**Exercício 4: (2 valores) Calcular para cada produto o total das vendas do ano e mostre-os usando Pie Chart (gráfico de queijo).**

```python
import pandas as pd
import matplotlib.pyplot as plt  

df = pd.read_csv("company_sales_data.csv")
monthList  = df ['month_number'].tolist() #criar lista

fig=plt.figure()
ax=fig.add_axes([0,0,1,1]) #aumentar os eixos a nível de largura e altura para uma melhor visualização do gráfico

labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer',] #criar as etiquetas que vão ficar no gráfico
salesData   = [df ['facecream'].sum(), df ['facewash'].sum(), df ['toothpaste'].sum(), 
         df ['bathingsoap'].sum(), df ['shampoo'].sum(), df ['moisturizer'].sum()] #soma dos valores de cada lista
plt.axis("equal") #para ficar um círculo perfeito
plt.pie(salesData, labels=labels, autopct='%1.1f%%') #o que o gráfico vai conter
plt.legend(loc='upper left') #onde ficam as legendas para não tapar o gráfico
plt.title('Sales data', fontsize= 17) #título do gráfico
plt.show()
```
