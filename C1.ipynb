# Exercício 1: 
<strong> Ler o lucro total (Total Profit) de todos os meses e mostre-os através
de um gráfico de linhas. </strong>

Usar o dado fornecido do lucro total para cada mês. A gráfico de linha gerado deve incluir
as seguintes propriedades:
X label name = Month Number
Y label name = Total profit

Cria as seguintes listas:
profitList = df ['total_profit'].tolist()
monthList = df ['month_number'].tolist()

```python
import pandas as pd
import matplotlib.pyplot as plt  

df = pd.read_csv("company_sales_data.csv") #importar os dados do csv
profitList = df ['total_profit'].tolist() #criar a lista profit que corresponde ao total profit 
monthList  = df ['month_number'].tolist() #criar a lista dos meses
plt.plot(monthList, profitList, label = 'Month-wise Profit data of last year') #armazenar na memória os tópicos do gráfico
plt.xlabel('Month number') #no eixo do x, criar a etiqueta do número de cada mês
plt.ylabel('Profit in dollar') #no eixo do y, criar a etiqueta do lucro em doláres
plt.xticks(monthList) #no eixo das coordenadas do x, são apresentada a lista dos meses
plt.title('Company profit per month') #título do gráfico
plt.yticks([100000,200000,300000,400000]) #no eixo das corrdenadas do y, são apresentados os valores do lucro/profit. De forma a simplicar, são apresentados apenas 4 valores
plt.show() #de forma a mostrar os conteúdos, armazenados na memória, do gráfico
```

# Exercício 2:
<strong> Ler o lucro total (Total Profit) de todos os meses e apresente-os num
gráfico de linhas, com as características indicadas. </strong>

A linha gerada deve incluir as seguintes propriedades de estilo:
-Estilo de linha pontilhado e cor da linha devem ser vermelhos
-Mostrar legenda no local inferior direito
-Label do rótulo X = Número do mês
-Label do rótulo Y = Número de unidades vendidas
-Adicione um marcador de círculo-Cor do marcador de linha conforme lida
-A largura da linha deve ser 3

Usar as listas:
profitList = df ['total_profit'].tolist()
monthList = df ['month_number'].tolist()

```python
import pandas as pd
import matplotlib.pyplot as plt  

df = pd.read_csv("company_sales_data.csv")
profitList = df ['total_profit'].tolist()
monthList  = df ['month_number'].tolist()

plt.plot(monthList, profitList, label = 'Profit data of last year', #armazenar na memória os tópicos do gráfico e a sua estética
      color='r', marker='o', markerfacecolor='k', #definir a cor como vermelha; marker definido como o; markerfacecolor-preto
      linestyle='--', linewidth=3)  #estilo e espessura da linha
      
plt.xlabel('Month Number')
plt.ylabel('Profit in dollar')
plt.legend(loc='lower right') #definir onde fica a legenda
plt.title('Company Sales data of last year')
plt.xticks(monthList)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

# Exercício 3:
<strong>Mostrar as vendas mensais de todos os produtos num gráfico de
múltiplas linhas, ou seja, uma linha diferente para cada produto.</strong>

```python
import pandas as pd
import matplotlib.pyplot as plt  

df = pd.read_csv("company_sales_data.csv")

#definir as listas de cada produto
monthList  = df ['month_number'].tolist()
faceCreamSalesData   = df ['facecream'].tolist()
faceWashSalesData   = df ['facewash'].tolist()
toothPasteSalesData = df ['toothpaste'].tolist()
bathingsoapSalesData   = df ['bathingsoap'].tolist()
shampooSalesData   = df ['shampoo'].tolist()
moisturizerSalesData = df ['moisturizer'].tolist()

#criar e personalizar a linha do gráfico de cada produto
plt.plot(monthList, faceCreamSalesData,   label = 'Face cream Sales Data', marker='o', linewidth=3)
plt.plot(monthList, faceWashSalesData,   label = 'Face Wash Sales Data',  marker='o', linewidth=3)
plt.plot(monthList, toothPasteSalesData, label = 'ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(monthList, bathingsoapSalesData, label = 'Bathing Soap Sales Data', marker='o', linewidth=3)
plt.plot(monthList, shampooSalesData, label = 'Shampoo Sales Data', marker='o', linewidth=3)
plt.plot(monthList, moisturizerSalesData, label = 'Moisturizer Sales Data', marker='o', linewidth=3)

#personalizar o gráfico de acordo com as coordenadas do eixo x e y (semelhante aos exercícios anteriores)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.xticks(monthList)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Sales data')
plt.show()
```

# Exercício 4:
<strong>Calcular para cada produto o total das vendas do ano e mostre-os
usando Pie Chart (gráfico de queijo).</strong>

No Pie chart inscrever a % de vendas de cada produto.

```python
import pandas as pd
import matplotlib.pyplot as plt  

df = pd.read_csv("company_sales_data.csv")
monthList  = df ['month_number'].tolist() #criar a lista dos meses

labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
salesData   = [df ['facecream'].sum(), df ['facewash'].sum(), df ['toothpaste'].sum(), 
         df ['bathingsoap'].sum(), df ['shampoo'].sum(), df ['moisturizer'].sum()]
plt.axis("equal")
plt.pie(salesData, labels=labels, autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.title('Sales data')
plt.show()
```
